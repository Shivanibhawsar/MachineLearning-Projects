## Plot Input Noisy Data

```python
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
plt.style.use('seaborn-whitegrid')
import numpy as np
noise_scale = 100
number_of_samples = 50
x = 25*(np.random.rand(number_of_samples, 1) - 0.5)
y = 5 * x + 15 * x**2 + 2 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
plt.plot(x,y,'ro')
```

## Plot Noisy Data & Polynomial with m = {1,2,3,4,5,6}

```python
x = x.flatten() #Flatten 2D data to 1D
m=[1,2,3,4,5,6]
mse=[] # To save Mean Squared Errors for different order m

for i in m:
    coefficients = np.polyfit(x, y, i)
    coefficients # In form of ..... + Ax^2 + Bx + C
    coefficients = coefficients.flatten() #Flatten 2D data to 1D
    polynomial = np.poly1d(coefficients)
    
    polynomial_x = np.linspace(np.min(x)-2, np.max(x)+2)
    polynomial_y = polynomial(polynomial_x)
    predicted_y = polynomial(x)
    plt.xlim([np.min(x)-2, np.max(x)+2 ])
    plt.plot(x, y, 'ro', polynomial_x,polynomial_y)
    plt.title('Polynomial Degree(m) = '+str(i))
    plt.show()
    
    mse_value = mean_squared_error(predicted_y, y);
    mse.append(mse_value)
```

## Plot MSE vs Order m where m = {1,2,3,4,5,6}

```python
plt.plot(m, mse)
print(mse)
plt.xlabel('m')
plt.ylabel('MSE')
plt.title("MSE vs m")
plt.show()
```

#### Based on above plot, if we increase m, MSE will decrease since polynomal curve with higher degree will fit our data. We can see that MSE for order m = 3,4,5 and 6 is constant and less than others and hence we can select m as 3,4,5 or 6
